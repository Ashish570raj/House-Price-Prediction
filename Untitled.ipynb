### Step 3.3: Drop features with too many missing values (> 50%)

```python
too_much_missing=missing_df[missing_df['Percentage']>50]['Feature'].tolist()
train.drop(columns=too_much_missing, inplace=True)
```

```python
print(train.head(20).isnull().sum())
```

```python
# Step 3.4: Fill categorical features with 'None'
none_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType',
             'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond',
             'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType']
for col in none_cols:
    if col in train.columns:
        train[col] = train[col].fillna('None')
```

```python
# Fill some categorical features with mode
mode_cols = ['Electrical', 'KitchenQual', 'Exterior1st', 'Exterior2nd', 'SaleType', 'MSZoning']

for col in mode_cols:
    if col in train.columns:
        train[col] = train[col].fillna(train[col].mode()[0])
```

```python
# Step 3.5: Fill numerical features with 0
zero_fill = ['GarageYrBlt', 'GarageArea', 'GarageCars', 'MasVnrArea',
             'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
             'BsmtFullBath', 'BsmtHalfBath']

for col in zero_fill:
    if col in train.columns:
        train[col] = train[col].fillna(0)
```

```python
# Fill LotFrontage using median of each neighborhood
if 'LotFrontage' in train.columns:
    train['LotFrontage'] = train.groupby('Neighborhood')['LotFrontage'].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
# Final check
print("âœ… Remaining missing values:", train.isnull().sum().sum())
```

## Feature Engineering

#### Log-Transform the Target Variable

```python
# Log transform the target to reduce skewness
train["SalePrice"] = np.log1p(train["SalePrice"])
```

 Why?
 
SalePrice is right-skewed. Log-transforming helps to normalize it for regression models.

#### Handling Skewed Numerical Features

```python
from scipy.stats import skew
```

```python
numeric_feats=train.select_dtypes(include=['int64','float64']).drop('SalePrice',axis=1)
```

```python
skewed_feats=numeric_feats.apply(lambda x:skew(x.dropna())).sort_values(ascending=False)
skewness=pd.DataFrame({'Skew':skewed_feats})
skewness = skewness[skewness['Skew'] > 0.75]
print(f"Skewed features:\n{skewness}")
```

```python
skewed_cols = skewness.index

for col in skewed_cols:
    train[col] = np.log1p(train[col])
    print(col)
```

```python
# Select some skewed features
features = ['LotArea', 'MiscVal', 'PoolArea']

# Plot histograms before and after log1p
for feature in features:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Original
    sns.histplot(train[feature].apply(np.expm1), kde=True, ax=axes[0], color='coral')
    axes[0].set_title(f'{feature} - Original')
    
    # After log transformation
    sns.histplot(train[feature], kde=True, ax=axes[1], color='teal')
    axes[1].set_title(f'{feature} - Log1p Transformed')

    plt.tight_layout()
    plt.show()
```
